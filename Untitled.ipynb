```python
import os
import random
import shutil
```

```python
# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏôÄ ÎùºÎ≤®Ïù¥ Ï†ÄÏû•Îêú Ìè¥Îçî Í≤ΩÎ°ú
images_dir = "/home/uit-lim/vspython/origin_files/images"
labels_dir = "/home/uit-lim/vspython/origin_files/labels"

print(images_dir)
print(labels_dir)
```

```python
# Î∂ÑÎ¶¨Îêú train, valid, test Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Ï†ÄÏû•Ìï† Ìè¥Îçî Í≤ΩÎ°ú
train_dir = "/home/uit-lim/vspython/change_files/train"
valid_dir = "/home/uit-lim/vspython/change_files/vaild"
test_dir = "/home/uit-lim/vspython/change_files/test"

print(train_dir)
```

```python
# train, valid, test ÎπÑÏú®
train_ratio = 0.8
valid_ratio = 0.1
test_ratio = 0.1
```

```python
# images Ìè¥Îçî ÎÇ¥Ïùò Î™®Îì† ÌååÏùºÏùÑ Î∂àÎü¨Ïò¥
images = os.listdir(images_dir)


# ÌååÏùºÏùÑ Î¨¥ÏûëÏúÑÎ°ú ÏÑûÏùå
random.shuffle(images)
```

```python
# train, valid, test Îç∞Ïù¥ÌÑ∞ÏÖãÏúºÎ°ú Î∂ÑÎ¶¨
train_images = images[:int(len(images) * train_ratio)]
valid_images = images[int(len(images) * train_ratio):int(len(images) * (train_ratio + valid_ratio))]
test_images = images[int(len(images) * (train_ratio + valid_ratio)):]
```

```python
# train, valid, test Ìè¥Îçî ÏÉùÏÑ±
os.makedirs(train_dir, exist_ok=True)
os.makedirs(valid_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
```

```python
# train Îç∞Ïù¥ÌÑ∞ÏÖã Ïù¥Îèô
for image in train_images:
    shutil.copy(os.path.join(images_dir, image), os.path.join(train_dir, image))
    shutil.copy(os.path.join(labels_dir, image.replace(".png", ".txt")), os.path.join(train_dir, image.replace(".png", ".txt")))
```

```python
# valid Îç∞Ïù¥ÌÑ∞ÏÖã Ïù¥Îèô
for image in valid_images:
    shutil.copy(os.path.join(images_dir, image), os.path.join(valid_dir, image))
    shutil.copy(os.path.join(labels_dir, image.replace(".png", ".txt")), os.path.join(valid_dir, image.replace(".png", ".txt")))
```

```python
# test Îç∞Ïù¥ÌÑ∞ÏÖã Ïù¥Îèô
for image in test_images:
    shutil.copy(os.path.join(images_dir, image), os.path.join(test_dir, image))
    shutil.copy(os.path.join(labels_dir, image.replace(".png", ".txt")), os.path.join(test_dir, image.replace(".png", ".txt")))
```
